```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
%matplotlib notebook
```

```python
x,y=make_regression(n_samples=1000,n_features=2,random_state=1,noise=10)# random state means that each time this will 
# produce same data
print(x.shape)
print(y.shape)
```

```python
plt.subplot(1,2,1)
plt.scatter(x[:,0],y)

plt.subplot(1,2,2)
plt.scatter(x[:,1],y)
plt.show()
```

```python
# 3D Visualization
from mpl_toolkits import mplot3d
fig=plt.figure(figsize=(10,6))
axes=plt.axes(projection="3d")
axes.scatter3D(x[:,0],x[:,1],y,cmap="rainbow",color="orange")
plt.title(" 3D scatter plot")
plt.show()
```

### Implementing Linear Regression using sklearn

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
# train the model
model.fit(x,y)
```

```python
model.coef_# are T1,T2 and so on
```

```python
model.intercept_# T0
```

```python
model.predict([x[0],x[1]])# return y^[0] and y^[1]
```

```python
y[0],y[1]
```

```python
model.score(x,y)
```
