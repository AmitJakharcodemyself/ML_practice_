```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.arange(10) # a ndarray 
Y=X**2+2 # function of X and Y-> error function 
print(X,Y)
```

- Our goal to minimize this error function and finding the value of x at ehich this is minimum

```python
# Visualization
plt.style.use("seaborn")
plt.plot(X,Y)
plt.ylabel("Error Function")
plt.xlabel("X or theta")
plt.show()
```

```python
# Gradient Descent Algo -> let's take or initialize the value of x=0,and ow we are changing value f x by x-=def(f(x)/dx
# by this we mean that we going down or stepping down and will stop where derivative of function =0-> local maxima or local minima
#we can also go for x+=d(f(x)/dx)-> stepping uo or going to upperside to the hill 
###
#in Gradient descent algo we check for local minima
###
```

```python
x=7# initial value of x and error at x=[7]# this is array 
lr=0.1# rate of stepping
# let's take 50 steps in downhill deirection as we know firstly we'll get local minima 
error=[]
for i in range(50):
    grad=2*x # deivative of error fun
    x=x-lr*grad
    e=x**2+2
    error.append(e)
    plt.scatter(x,e)
    print(x)
```

```python
# Plot tha values of error
plt.plot(error)
plt.show()
```
