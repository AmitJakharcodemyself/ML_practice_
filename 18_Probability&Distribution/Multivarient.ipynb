## Multivarient Normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# bivariant-> m==2
mean=np.array([0.0,0.0])
cov=np.array([[1,0],[0,1]])
dist=np.random.multivariate_normal(mean,cov,500)
print(dist)
print(dist.shape)
print(type(dist))# np array 
```

```python
plt.scatter(dist[:,0],dist[:,1])# x-> column[0] and y-> column[1]
plt.show()
```
