## Histogram

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# to generate a random data having mean=0 and std(standard deviation =1)=1 # it is Std. Normal Distribuyion
Xsn=np.random.randn(100) # 100 denotes no of point we want 
```

```python
print(Xsn)
```

```python
# to generate random data distribution having mean value=60 and sigma=5
sigma=5
u=60
X=np.round(Xsn*sigma + u)
print(X)
X1=np.round(Xsn*6+20)
```

```python
plt.style.use('seaborn')
```

```python
plt.hist(X,alpha=0.8,label="maths")
plt.hist(X1,alpha=0.9,label="physics")
plt.xlabel("marks")
plt.ylabel("prob/freq")
plt.legend()
plt.title("Histogram")
plt.show()
```
