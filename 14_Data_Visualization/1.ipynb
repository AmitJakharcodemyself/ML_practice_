```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
themes=plt.style.available
print(themes)
```

```python
plt.style.use('seaborn')
```

```python
x=np.arange(10)
y1=x**2
y2=2*x +5
print(x)
print(y1)
print(y2)
```

```python
plt.plot(x,y1,color='red',label='Apple',marker='o')
plt.plot(x,y2,color='green',label='Mango',marker='*',linestyle='dashed')
plt.xlabel('time')
plt.ylabel('price')
plt.title('Prices of fruits over the time')
plt.legend()
plt.show()
```

```python
# if we give only one array as in argument then it' consider it as y and assume indexes as x array
prices=np.array([1,2,3,4])**2
plt.plot(prices)
plt.show()
```

## Scatter Plot

```python
# Adjust size of the plot
#plt.figure(figsize=(5,6))
plt.figure(figsize=(10,10))
plt.scatter(x,y1,color='red',label='Apple',marker='^')
plt.scatter(x,y2,color='green',label='Mango',marker='*',linestyle='dashed')
plt.xlabel('time')
plt.ylabel('price')
plt.title(' Scatter-Plot:Prices of fruits over the time')
plt.legend()
plt.show()
```

## Bar-Graphs

```python
x_coordinates=np.array([0,1,2,3])*2
tick_labels=["gold","platinum","silver","Aluminium"]
# we can provide array or list in bar 
plt.bar(x_coordinates-0.25,[10,15,20,25],width=0.5,label='current year',tick_label=tick_labels)# current year
plt.bar(x_coordinates+0.25,[14,20,12,22],width=0.5,label="next year")# next year
plt.ylim(0,40)
plt.xlim(-2,5)
plt.xlabel("Metal")
plt.ylabel("price")
plt.title("Metal prices over 2 years")
plt.legend()
plt.show()

```

```python
#plt.style.use('dark_background')
x_coordinates=np.array([0,1,2,3])*2
tick_labels=["gold","platinum","silver","Aluminium"]
# we can provide array or list in bar 
plt.bar(x_coordinates-0.25,[10,15,20,25],width=0.5,label='current year',tick_label=tick_labels,color='orange')# current year
plt.bar(x_coordinates+0.25,[14,20,12,22],width=0.5,label="next year",color='green')# next year
plt.ylim(0,40)
plt.xlim(-2,5)
plt.xlabel("Metal")
plt.ylabel("price")
plt.title("Metal prices over 2 years")
plt.legend()
plt.show()
```
